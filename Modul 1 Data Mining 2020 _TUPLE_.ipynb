Sebuah tuple adalah **sequences Python object yang Immutable.** Tuple menggunakan tanda kurung `()` untuk mendefinisikannya.

Sebagai contoh berikut cara membuat tuple :

```
tup1 = ('physics', 'chemistry', 1997, 2000)
tup2 = (1, 2, 3, 4, 5 )
tup3 = "a", "b", "c", "d"
tup4 = ()
```

Gunakan method `tuple()` untuk mengubah list ke tuple

```
lst3  = ["a", "b", "c", "d"]
tup6 = tuple(lst3)
tup6
```

Method `type()` dapat digunakan untuk mengetahui jenis tipe data dari suatu data.

```
tup5 =(50,)
print(type(tup5))

tup5 =(50)
print(type(tup5))

tup5 =("50")
print(type(tup5))
```

**NOTE :**

*Jika ingin mendekarasikan **tuple dengan single element** maka selalu **sertakan tanda koma (,) di belakang** untuk mengindikasikan bahwa itu adalah tuple bukan tipe data lain.*

### **Updating Tuples**

Tuples bersifat immutable, yang berarti **Anda tidak dapat meng update atau mengubah nilai element tuple**. Anda dapat mengambil element tupel yang ada untuk membuat tupel baru seperti yang diperlihatkan contoh berikut :

```
tup7 = (12, 34.56)
tup8 = ('abc', 'xyz')

tup9 = tup7 + tup8
print (tup9)
```

### **Delete Tuple Elements**

Menghapus elemen tuple individual tidak dimungkinkan. Tentu saja, tidak ada yang salah dengan menyusun tuple lain dengan elemen yang tidak menghilangakan diinginkan.

Untuk menghapus seluruh tuple secara eksplisit, cukup gunakan statement del. Sebagai contoh :

```
tup = ('physics', 'chemistry', 1997, 2000);
print (tup)
del tup;
```

```
print ("After deleting tup : ")
print (tup)
```

### **Basic Tuples Operations**

Tuples respond to the + and * operators much like strings; they mean concatenation and repetition here too, except that the result is a new tuple, not a string.


```
len((1, 2, 3))	# Length
```

```
(1, 2, 3) + (4, 5, 6)	# Concatenation
```

```
('Hi!',) * 4	#	Repetition
```
