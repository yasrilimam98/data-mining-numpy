## **1D Numpy Array**

Array NumPy mirip dengan list. Biasanya **ukurannya tetap** dan setiap **elemen bertipe sama.** Untuk membuat array NumPy kita terlebih dahulu harus mengimpor NumPy:

```
import numpy as np 
```

```
x = [1, 2, 3, 4, 5]
x
```

```
a = np.array(x)
a
```

```
type(a)
```

```
type(x)
```

### **Assign value**

Kita dapat mengubah nilai dari np array perhatikan contoh berikut :

```
print("Before Updating",a)
a[0]=100
print("After Updating",a)
```

### **Slicing**

Kita dapat membuat np array baru yang berisi beberapa anggota elemen dari np array lainnya. Cukup gunakan `[:]` untuk melakukan slicing di ikuti index tujuan.

```
b = a[1:4]
print(b)
b = a[-4:-2]
print(b)
b = a[:3]
print(b)
b = a[2:]
print(b)
b = a[:]
print(b)
```

Dengan menggunakan slicing kita juga bisa mengubah isi np array sesuai keinginan kita perhatikan code berikut:

```
a[2:4] = 300, 400
a
```

Anda juga dapat mengakses dan menggubah nilai np array menggunakan list.

```
my_list = [0, 2, 3]
c = a[my_list]
c
```

```
a[my_list] = 5000
a
```

### **NumPy Array Operations**

**Basic Operations**

```
# Dapatkan jumlah elemen dalam array:
print(a.size)

# ndim digunakan untuk mendapatkan jumlah sumbu atau dimensi yang disebut pangkat:
print(a.ndim)

# Atribut "shape" adalah tupel bilangan bulat yang menunjukkan ukuran array di setiap dimensi:
print(a.shape)

# Dapatkan nilai mean dari elemen numpy array:
print(a.mean())

# Dapatkan standar deviasi dari elemen numpy array:
print(a.std())

# Dapatkan nilai terbesar dari numpy array:
print(a.max())

# Dapatkan nilai terkecil dari numpy array:
print(a.min())
```

**Simple Math Operations**

```
u = np.array([1, 2, 3, 4])
v = np.array([3, 1, 4, 2])
```

```
# Kita dapat menambahkan dua array dan menetapkannya ke z:
z = u + v
print("Penambahan dua array",z)

# Hitung produk dari dua array numpy:
z = u * v
print("Perkalian dua array",z)

# Hitung dot product:
print("Dot product matrix",np.dot(u, v))

# Kita bisa mengalikan setiap elemen dalam array dengan konstanta:
z = 2 * u
print("Perkalian array dengan bilangan bulat",z)

z = 2.3 * u
print("Perkalian array dengan bilangan desimal",z)

# Tambahkan konstanta ke array:
u = u + 5
print("Penambahan array dengan konstanta",u)

# Pengurangan konstanta ke array:
u = u - 5
print("Pengurangan array dengan konstanta",u)
```

### **Linspace**

Fungsi yang berguna untuk lainnya adalah "linespace". Linespace mengembalikan angka yang diberi spasi secara merata selama interval yang ditentukan. Kami menentukan titik awal dari urutan dan titik akhir dari urutan. Parameter "num" menunjukkan Jumlah sampel yang akan dihasilkan:

```
# Membuat array dengan 9 elemen dari rentang [-2, 4]
np.linspace(-2, 4, num=9)
```
